```python
import pandas as pd
import numpy as np
import lightgbm as lgb
import gc
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.preprocessing import StandardScaler
import joblib

np.random.seed(2020)


def lgb_model_age(train_x, test_x,train_y):
    
    model = joblib.load('w2v_lgb_age_2.pkl')
    
    y_t_pred = model.predict(test_x)
    res['predicted_age']=y_t_pred

    # 评估训练样本的性能指标
    train_pred = model.predict(train_x)
    # print(model.get_score(importance_type='weight'))
    cm = confusion_matrix(train_y, train_pred)
    np.set_printoptions(precision=3)                                    # 显示精度
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]  # 将样本矩阵转化为比率

    print('age************************************')
    print('confusion_matrix is \n {:} \n '.format(cm_normalized))
    print('test acc is \n {:} \n '.format(np.sum(train_y==train_pred)/len(train_y)))
    print(classification_report(train_y,train_pred))
    print('accuracy is %f , sen is %f,spe is %f ' % (accuracy_score(train_y, train_pred) * 100,
                                                     cm_normalized[0][0],cm_normalized[1][1] ))

    return accuracy_score(train_y, train_pred)

def lgb_model_gender(train_x, test_x,train_y):
    
    model = joblib.load('w2v_lgb_gender_2.pkl')
    
    y_t_pred = model.predict(test_x)
    res['predicted_gender'] = y_t_pred

    # 评估训练样本的性能指标
    train_pred = model.predict(train_x)
    # print(model.get_score(importance_type='weight'))
    cm = confusion_matrix(train_y, train_pred)
    np.set_printoptions(precision=3)  # 显示精度
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]  # 将样本矩阵转化为比率

    print('gender************************************')
    print('confusion_matrix is \n {:} \n '.format(cm_normalized))
    print('test acc is \n {:} \n '.format(np.sum(train_y == train_pred) / len(train_y)))
    print(classification_report(train_y, train_pred))
    print('accuracy is %f , sen is %f,spe is %f ' % (accuracy_score(train_y, train_pred) * 100,
                                                     cm_normalized[0][0], cm_normalized[1][1]))
    return accuracy_score(train_y, train_pred)

def load_data():
    # user
    train_data = pd.read_csv('../w2v_feat_data/train_data.csv')
    train_label = train_data[['age','gender']]
    train_data = train_data.drop(['user_id','age','gender'],axis = 1)
    
    test_data = pd.read_csv('../w2v_feat_data/test_data.csv')
    res = test_data[['user_id']]
    test_data = test_data.drop(['user_id'],axis = 1)
    
    return train_data,train_label,test_data,res


train_data,train_label,test_data,res = load_data()

# stand = StandardScaler()
# stand.fit(train_age_x)
# train_age_x, test_age_x = stand.transform(train_age_x),stand.transform(test_age_x)
# 训练模型
acc_age = lgb_model_age(train_data, test_data,train_label['age'])


# 训练模型
acc_gender = lgb_model_gender(train_data, test_data,train_label['gender'])
print(acc_age+acc_gender)

res.to_csv('submission.csv',index=False)                          # submission.csv
```
